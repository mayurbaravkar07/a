```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Input data
X = np.array([[7, 8], [6, 9], [8, 6], [8, 7], [9, 5], [5, 7]])
y = np.array([1, 1, -1, -1, -1, 1])
```

```python
# Initialize weights randomly
w = np.random.rand(2)
```

```python

# Define the perceptron function
def perceptron(x, w):
    return np.dot(x, w)
```

```python

# Define the learning rate
eta = 0.1
```

```python

# Define the number of epochs
epochs = 10

# The sign function provided by the NumPy library. 
#The sign function returns -1 for negative values, 
#0 for zero, and 1 for positive values. 
#It essentially determines the sign of the input value.
```

```python

# Loop through the epochs
for epoch in range(epochs):
    # Loop through each data point
    for i in range(len(X)):
        # Compute the predicted label
        y_pred = np.sign(perceptron(X[i], w)) 
        print(y_pred)
        # Update the weights 
        #if the predicted label is different from the true label
        if y_pred != y[i]:
            w = w + eta * y[i] * X[i]
            
            
    # Plot the data points and the decision boundary
    plt.scatter(X[:,0], X[:,1], c=y)
    x_boundary = np.array([np.min(X[:,0]), np.max(X[:,0])])
    y_boundary = -(w[0]/w[1]) * x_boundary
    plt.plot(x_boundary, y_boundary)
    plt.show()
```
