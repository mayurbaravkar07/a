```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp
```

```python
#1. Sigmoid Activation Function
```

```python
def Stepfun(x):
    return np.array(x > 0 , dtype = int)

inputs = [x for x in range(-50,50)]
outputs = [Stepfun(x) for x in inputs]
plt.title("1.Binary Step Function")
plt.plot(inputs,outputs)
plt.show()
```

```python
# 2.ReLU Activation Funciton
```

```python
def Relu(x):
    return max(0,x)
inputs = [x for x in range(-10,10)]
outputs = [Relu(x) for x in inputs]
plt.title("2.ReLU activation Function")
plt.plot(inputs, outputs)
plt.show()
```

```python
def Sigmoid(x):
    return 1/(1+np.exp(-x))

inputs = [x for x in range(-10,10)]
outputs = [Sigmoid(x) for x in inputs]
plt.title("3.Sigmoid Activation Function")
plt.plot(inputs,outputs)
plt.show()
```

```python
# 4.Tanh Activation Function
```

```python
def Tanh(x):
    return (np.exp(x) - np.exp(-x))/ (np.exp(x) + np.exp(-x))

inputs = [x for x in range(-10,10)]
outputs = [Tanh(x) for x in inputs]
plt.title("4. Tanh Activation Function")
plt.plot(inputs, outputs)
plt.show()
```

```python
#5.Linear Activation Function
```

```python
def Linear(x):
    return x

inputs = [x for x in range(-10,10)]
outputs = [Linear(x) for x in inputs]
plt.title("5.Linear Activation Function")
plt.plot(inputs , outputs)
plt.show()
```

```python
#6.Leaky ReLU
```

```python
def LeakRelu(x , alpha = 0.1):
    return np.maximum(alpha*x , x)

inputs = [x for x in range(-10,10)]
outputs = [LeakRelu(x) for x in inputs]
plt.title("6.Leaky ReLU Activation Function")
plt.plot(inputs , outputs)
plt.show()
```
